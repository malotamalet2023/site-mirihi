## 2. Configuration API et Données de Test

Configuration de l'URL de l'API locale et création de données de diagnostic simulées

```python
# Configuration de l'API
API_BASE_URL = "http://localhost:3001"
ORIENTATION_API_ENDPOINT = f"{API_BASE_URL}/api/orientation-analysis"

print(f"🔗 API Endpoint: {ORIENTATION_API_ENDPOINT}")

# Données de diagnostic simulées - Profil débutant avec faiblesses
diagnostic_data_weak = {
    "overallScore": 35,
    "categoryScores": {
        "Engagement Direction & Stratégie": {"score": 8, "maxScore": 25, "percentage": 32, "level": "faible"},
        "Relations avec Partenaires Internes": {"score": 10, "maxScore": 20, "percentage": 50, "level": "moyen"},
        "Gestion par Catégories": {"score": 6, "maxScore": 20, "percentage": 30, "level": "faible"},
        "Collaboration et Innovation Fournisseurs": {"score": 5, "maxScore": 15, "percentage": 33, "level": "faible"},
        "Digitalisation des Achats": {"score": 6, "maxScore": 20, "percentage": 30, "level": "faible"}
    },
    "answers": [],
    "timestamp": datetime.now().isoformat()
}

# Données de diagnostic simulées - Profil avancé avec quelques points à améliorer
diagnostic_data_advanced = {
    "overallScore": 78,
    "categoryScores": {
        "Engagement Direction & Stratégie": {"score": 20, "maxScore": 25, "percentage": 80, "level": "bon"},
        "Relations avec Partenaires Internes": {"score": 16, "maxScore": 20, "percentage": 80, "level": "bon"},
        "Gestion par Catégories": {"score": 15, "maxScore": 20, "percentage": 75, "level": "bon"},
        "Collaboration et Innovation Fournisseurs": {"score": 12, "maxScore": 15, "percentage": 80, "level": "bon"},
        "Digitalisation des Achats": {"score": 15, "maxScore": 20, "percentage": 75, "level": "bon"}
    },
    "answers": [],
    "timestamp": datetime.now().isoformat()
}

print("✅ Données de test créées :")
print(f"   - Profil débutant : {diagnostic_data_weak['overallScore']}%")
print(f"   - Profil avancé : {diagnostic_data_advanced['overallScore']}%")
```

## 3. Test de l'API d'Analyse d'Orientation

Test de l'API avec Gemini 2.5 Flash et Google Search pour analyser différents profils

```python
def test_orientation_api(diagnostic_data, profile_name):
    """
    Teste l'API d'orientation avec des données de diagnostic
    """
    print(f"\n🧪 Test de l'API pour le profil : {profile_name}")
    print(f"📊 Score global : {diagnostic_data['overallScore']}%")
    print("-" * 50)
    
    try:
        # Préparation de la requête
        payload = {
            "diagnosticData": diagnostic_data
        }
        
        headers = {
            "Content-Type": "application/json"
        }
        
        print("📤 Envoi de la requête à l'API...")
        
        # Appel de l'API
        response = requests.post(
            ORIENTATION_API_ENDPOINT,
            json=payload,
            headers=headers,
            timeout=30
        )
        
        print(f"📥 Réponse reçue - Status Code: {response.status_code}")
        
        if response.status_code == 200:
            result = response.json()
            print("✅ Succès ! Analyse reçue")
            
            # Affichage des résultats
            if result.get("success"):
                print(f"\n🤖 Insights Gemini avec Google Search :")
                insights = result.get("insights", "Non disponible")
                print(f"   {insights[:200]}..." if len(insights) > 200 else f"   {insights}")
                
                print(f"\n💡 Recommandations ({len(result.get('recommendations', []))}) :")
                for i, rec in enumerate(result.get("recommendations", [])[:3], 1):
                    print(f"   {i}. {rec}")
                
                print(f"\n🎯 Prochaines étapes ({len(result.get('nextSteps', []))}) :")
                for i, step in enumerate(result.get("nextSteps", [])[:3], 1):
                    print(f"   {i}. {step}")
                
                print(f"\n📈 Benchmark industrie :")
                benchmark = result.get("industryBenchmark", "Non disponible")
                print(f"   {benchmark}")
                
                return result
            else:
                print(f"❌ Erreur dans la réponse : {result.get('error', 'Inconnue')}")
                return None
        else:
            print(f"❌ Erreur HTTP {response.status_code}: {response.text}")
            return None
            
    except requests.exceptions.Timeout:
        print("⏰ Timeout - L'API met trop de temps à répondre")
        return None
    except requests.exceptions.ConnectionError:
        print("🔌 Erreur de connexion - Vérifiez que le serveur est lancé")
        return None
    except Exception as e:
        print(f"💥 Erreur inattendue : {str(e)}")
        return None

print("✅ Fonction de test définie")
```

```python
# Test avec le profil débutant
print("🚀 Lancement du test - Profil Débutant")
result_weak = test_orientation_api(diagnostic_data_weak, "Profil Débutant (35%)")
```

```python
# Test avec le profil avancé
print("\n" + "="*60)
print("🚀 Lancement du test - Profil Avancé")
result_advanced = test_orientation_api(diagnostic_data_advanced, "Profil Avancé (78%)")
```
